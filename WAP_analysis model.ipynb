```python
import os
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("E:/AII/AII_Course-1-main/Classification/wifi_signal_strength/data/UJIndoorLoc/trainingData.csv")
train.head()
```

```python
print(train.isnull().sum())
```

```python
train["BUILDINGID"].unique()
```

```python
train["FLOOR"].unique()
```

```python
train['BUILDINGID'] = train['BUILDINGID'].map(str)
train['FLOOR'] = train['FLOOR'].map(str)

train['LONGITUDE'] = train['LONGITUDE'].map(str)
train['LATITUDE'] = train['LATITUDE'].map(str)
```

```python
train["FLOOR_ID"] = train[["BUILDINGID", "FLOOR"]].apply(lambda x: "-".join(x), axis =1)

train["LOCATION"] = train[["LONGITUDE", "LATITUDE"]].apply(lambda x: "-".join(x), axis =1)
```

```python
train["FLOOR_ID"].unique()
```

```python
drop_col = ['FLOOR','BUILDINGID','USERID', 'TIMESTAMP','SPACEID','RELATIVEPOSITION','LONGITUDE','LATITUDE','SPACEID', 'RELATIVEPOSITION']
train = train.drop(drop_col, axis=1)
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
train['FLOOR_ID'] = label_encoder.fit_transform(train['FLOOR_ID'])
train['LOCATION'] = label_encoder.fit_transform(train['LOCATION'])

train.head()
```

```python
y_train = train['LOCATION']
y_train
```

```python
X_train = train.drop(['LOCATION'], axis=1)
X_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier

from sklearn.model_selection import cross_val_score
```

```python
# Define the models for evaluation
models = [
    ('Decision Tree', DecisionTreeClassifier()),
    ('Random Forest', RandomForestClassifier()),
    ('Naive Bayes', GaussianNB()),
    ('Neural Network', MLPClassifier(max_iter=10000)),
]
```

```python
for name, model in models:
    scores = cross_val_score(model, X_train, y_train, cv=5)
    print(f"{name}: Accuracy = {scores.mean()}, Standard Deviation = {scores.std()}")
```
